```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.edge.options import Options
from webdriver_manager.microsoft import EdgeChromiumDriverManager
import pandas as pd
import time

# Configure Edge WebDriver
edge_options = Options()
edge_options.add_argument("--disable-gpu")
edge_options.add_argument("--window-size=1920,1080")
edge_options.add_argument("--disable-blink-features=AutomationControlled")
# REMOVE HEADLESS MODE FOR DEBUGGING
# edge_options.add_argument("--headless")  

# Initialize WebDriver
service = Service(EdgeChromiumDriverManager().install())
driver = webdriver.Edge(service=service, options=edge_options)

City = "Gurgaon"
start = int(input("Enter the page number to start from: "))
end = start + 5  # Fetch 5 pages for testing
pageNumber = start
flats = pd.DataFrame()

try:
    while pageNumber < end:
        url = f"https://www.99acres.com/flats-in-{City}-ffid-page-{pageNumber}"
        print(f"\nFetching: {url}")

        try:
            driver.get(url)
            # Wait until listings load (Increase timeout if needed)
            WebDriverWait(driver, 15).until(
                EC.presence_of_element_located((By.CSS_SELECTOR, 'section[data-hydration-on-demand="true"]'))
            )

            listings = driver.find_elements(By.CSS_SELECTOR, 'section[data-hydration-on-demand="true"]')
            
            if not listings:
                print(f"⚠️ No listings found on page {pageNumber}. Skipping...")
                pageNumber += 1
                continue
            
            for listing in listings:
                try:
                    # Property Name
                    name_elem = listing.find_element(By.CSS_SELECTOR, 'a.srpTuple__propertyName')
                    property_name = name_elem.text.strip()
                    link = name_elem.get_attribute('href')

                    # Society Name
                    try:
                        society_elem = listing.find_element(By.CSS_SELECTOR, '[id*="society_heading"]')
                        society = society_elem.text.strip()
                    except:
                        society = ''

                    # Area
                    try:
                        area_elem = listing.find_element(By.CSS_SELECTOR, '[id*="price_per_unit_area"]')
                        area = area_elem.text.strip()
                    except:
                        area = ''

                    # Price
                    try:
                        price_elem = listing.find_element(By.CSS_SELECTOR, '[id*="pdPrice2"]')
                        price = price_elem.text.strip()
                    except:
                        price = ''

                    property_data = {
                        'property_name': property_name,
                        'link': link,
                        'society': society,
                        'area': area,
                        'price': price
                    }

                    temp_df = pd.DataFrame([property_data])
                    flats = pd.concat([flats, temp_df], ignore_index=True)

                except Exception as e:
                    print(f"Error processing a listing: {e}")

            print(f"✅ Page {pageNumber} processed.")
            pageNumber += 1
            time.sleep(5)  # Reduce request rate

        except Exception as e:
            print(f"Error fetching {url}: {e}")
            print("Retrying in 20 seconds...")
            time.sleep(20)

except Exception as e:
    print(f"Unexpected error: {e}")

# Save Data
csv_file_path = f"flats_{City}_data-page-{start}-{pageNumber-1}.csv"
flats.to_csv(csv_file_path, index=False)
print(f"Data saved to {csv_file_path}")

# Close browser
driver.quit()
```

```python
# Function to combine multiple csv file is one file.

def combine_csv_files(folder_path, combined_file_path):
    combined_data = pd.DataFrame()  # Create an empty DataFrame to hold the combined data

    # Iterate through all CSV files in the folder
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.csv'):
            file_path = os.path.join(folder_path, file_name)
            print('file_path')
            # Read the data from the current CSV file
            df = pd.read_csv(file_path)

            # Append the data to the combined DataFrame
            combined_data = combined_data.append(df, ignore_index=True)

            # Delete the original CSV file
            os.remove(file_path)

    # Save the combined data to a new CSV file
    combined_data.to_csv(combined_file_path, index=False)

# Example usage:

# Replace with the actual folder path
folder_path = '/content/drive/MyDrive/DSMP/Case Studies/Real estate/flats_appartment'

# Replace with the desired combined file path
combined_file_path = '/content/drive/MyDrive/DSMP/Case Studies/Real estate/flats_appartment/flats.csv'

combine_csv_files(folder_path, combined_file_path)
```
